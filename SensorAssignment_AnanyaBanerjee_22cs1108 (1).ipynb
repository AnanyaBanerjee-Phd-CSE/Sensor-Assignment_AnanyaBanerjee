```python

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/sitting.csv')

# Create a new figure with 4 subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 15))

# Define line styles and line widths for distinction
line_styles = ['--', '-', '-']
marker_styles = [None, '+', None]
line_widths = [2, 4.1, 5.2]
colors = ['red', 'cyan', 'green']

# Plotting the LSM6DSO accelerometer data
for i, axis in enumerate(['x', 'y', 'z']):
    axs[0].plot(data[f'lsm6dso_accelerometer.{axis}'], label=f'{axis.upper()}-axis', linestyle=line_styles[i], marker=marker_styles[i], linewidth=line_widths[i], color=colors[i])
axs[0].set_title('LSM6DSO Accelerometer')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Acceleration (m/s^2)')
axs[0].legend()

# Plotting the LSM6DSO gyroscope data
for i, axis in enumerate(['x', 'y', 'z']):
    axs[1].plot(data[f'lsm6dso_gyroscope.{axis}'], label=f'{axis.upper()}-axis', linestyle=line_styles[i], marker=marker_styles[i], linewidth=line_widths[i], color=colors[i])
axs[1].set_title('LSM6DSO Gyroscope')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Angular Velocity (/s)')
axs[1].legend()

# Plotting the TMD4912 light data
for i, axis in enumerate(['x', 'y', 'z']):
    axs[2].plot(data[f'tmd4912_light_cct.{axis}'], label=f'{axis.upper()}-axis', linestyle=line_styles[i], marker=marker_styles[i], linewidth=line_widths[i], color=colors[i])
axs[2].set_title('TMD4912 Light')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Illuminance (lx)')
axs[2].legend()

# Plotting the AK09918 magnetometer data
for i, axis in enumerate(['x', 'y', 'z']):
    axs[3].plot(data[f'ak09918_magnetometer.{axis}'], label=f'{axis.upper()}-axis', linestyle=line_styles[i], marker=marker_styles[i], linewidth=line_widths[i], color=colors[i])
axs[3].set_title('AK09918 Magnetometer')
axs[3].set_xlabel('Time')
axs[3].set_ylabel('Orientation ()')
axs[3].legend()

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()


















```
