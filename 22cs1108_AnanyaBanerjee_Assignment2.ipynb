```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV file
df = pd.read_csv('/content/AC Classroom.csv')

# Assuming the data was collected at intervals of 5 minutes for an hour
time_intervals = [f'{i*5} min' for i in range(12)]
df['Time'] = time_intervals

# Melt the dataframe to have a format suitable for seaborn's barplot
df_melted = df.melt(id_vars="Time",
                    value_vars=['Temp', 'Humidity', 'PM2.5', 'PM10'],
                    var_name="Parameter",
                    value_name="Value")

# Create a custom color palette for the column chart
custom_palette = sns.color_palette("Set2")

# Create a column chart (vertical bar plot) with customized y-axis label
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(121)  # Create the first subplot (column chart)
sns.barplot(x="Time", y="Value", hue="Parameter", data=df_melted, palette=custom_palette)

plt.title('Column Chart for Particulate Matter Over Time Intervals')
plt.ylabel('Particulate Matter (µg/m³)')
plt.xticks(rotation=45)
plt.legend(title='Parameters')

# Sum the values for each parameter over all time intervals for the pie chart
sum_values = df[['Temp', 'Humidity', 'PM2.5', 'PM10']].sum()
labels = sum_values.index
values = sum_values.values

# Create the pie chart
ax2 = plt.subplot(122)  # Create the second subplot (pie chart)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#c2c2f0']
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, shadow=True)
plt.title('Pie Chart for Particulate Matter Distribution')

# Show both charts
plt.tight_layout()
plt.show()
```
